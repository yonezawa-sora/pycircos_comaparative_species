```python
import pandas as pd
```

```python
# Upregulated genes
df_up = pd.read_csv("../data/HN-score_rice/HN5_genelist_rice_2311/HN5_genes_up_rice.tsv", sep="\t")
df_up.rename(columns={'HN5': 'HN-score(HN5)'}, inplace=True)

# Downregulated genes
df_down = pd.read_csv("../data/HN-score_rice/HN5_genelist_rice_2311/HN5_genes_down_rice.tsv", sep="\t")
df_down.rename(columns={'HN5': 'HN-score(HN5)'}, inplace=True)

# Combine
combined = pd.concat(
    [df_up, df_down],
    ignore_index=True
)

display(df_up)
display(df_down)
display(combined)
```

## Merge the rice gene positoins and the rice gene scores

```python
df_ensembl_position = pd.read_csv("../data/biomart_goslim/biomart_rice_position_R56.tsv", sep="\t")

# Merge all
merge_df_position = pd.merge(
    df_ensembl_position,
    combined,
    left_on='Gene stable ID',
    right_on='GENEID',
    how='inner'
)
merge_df_position = merge_df_position.drop(columns=['GENEID'])
merge_df_position = merge_df_position.sort_values(by='HN-score(HN5)', ascending=False).copy()
merge_df_position.to_csv("../data/rice_annotation/rice_position.tsv", sep="\t", index=False)

# merge up-regulated genes
merge_df_position_up = pd.merge(
    df_ensembl_position,
    df_up,
    left_on='Gene stable ID',
    right_on='GENEID',
    how='inner'
)
merge_df_position_up = merge_df_position_up.drop(columns=['GENEID'])
merge_df_position_up = merge_df_position_up.sort_values(by='HN-score(HN5)', ascending=False).copy()
merge_df_position_up.to_csv("../data/rice_annotation/rice_position_up.tsv", sep="\t", index=False)

# merge down-regulated genes
merge_df_position_down = pd.merge(
    df_ensembl_position,
    df_down,
    left_on='Gene stable ID',
    right_on='GENEID',
    how='inner'
)
merge_df_position_down = merge_df_position_down.drop(columns=['GENEID'])
merge_df_position_down = merge_df_position_down.sort_values(by='HN-score(HN5)', ascending=False).copy()
merge_df_position_down.to_csv("../data/rice_annotation/rice_position_down.tsv", sep="\t", index=False)


display(df_ensembl_position)
display(merge_df_position)
display(merge_df_position_up)
display(merge_df_position_down)
```

## merge the Ensembl biomart data (GO slim)

```python
df_ensembl = pd.read_csv("../data/biomart_goslim/biomart_rice_goslim_R56.tsv", sep="\t")
df_ensembl_uniq = df_ensembl.drop_duplicates(subset=['Gene stable ID', 'GOSlim GOA Accession(s)', 'GOSlim GOA Description'], keep='first').copy()
df_ensembl_uniq.rename(columns={'start1 (bp)': 'Gene start(bp)'}, inplace=True)
df_ensembl_uniq.rename(columns={'end1 (bp)': 'Gene end(bp)'}, inplace=True)

# 1. merge all data
merged_df = pd.merge(df_ensembl_uniq,combined,
                        left_on="Gene stable ID",
                        right_on="GENEID",
                     how="inner")

merged_df = merged_df.drop(columns=['GENEID'])
merged_df = merged_df.sort_values(by='HN-score(HN5)', ascending=False).copy()
merged_df.to_csv("../data/rice_annotation/rice_annotation.tsv", sep="\t", index=False)

# 2. merge UP data
merged_df_up = pd.merge(df_ensembl_uniq,df_up,
                        left_on="Gene stable ID",
                        right_on="GENEID",
                     how="inner")

merged_df_up = merged_df_up.drop(columns=['GENEID'])
merged_df_up = merged_df_up.sort_values(by='HN-score(HN5)', ascending=False).copy()
merged_df_up.to_csv("../data/rice_annotation/rice_annotation_up.tsv", sep="\t", index=False)


# 3. merge DOWN data
merged_df_down = pd.merge(df_ensembl_uniq,df_down,
                        left_on="Gene stable ID",
                        right_on="GENEID",
                     how="inner")

merged_df_down = merged_df_down.drop(columns=['GENEID'])
merged_df_down = merged_df_down.sort_values(by='HN-score(HN5)', ascending=False).copy()
merged_df_down.to_csv("../data/rice_annotation/rice_annotation_down.tsv", sep="\t", index=False)


display(df_ensembl_uniq)
display(merged_df)
display(merged_df_up)
display(merged_df_down)
```
