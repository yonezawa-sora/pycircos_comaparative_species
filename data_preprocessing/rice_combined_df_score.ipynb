```python
import pandas as pd
```

```python
# Upregulated genes
df_up = pd.read_csv("../data/HN-score_rice/HN5_genelist_rice_2311/HN5_genes_up_rice.tsv", sep="\t")
df_up.rename(columns={'HN5': 'HN-score(HN5)'}, inplace=True)
df_up
```

```python
# Downregulated genes
df_down = pd.read_csv("../data/HN-score_rice/HN5_genelist_rice_2311/HN5_genes_down_rice.tsv", sep="\t")
df_down.rename(columns={'HN5': 'HN-score(HN5)'}, inplace=True)
df_down
```

```python
combined = pd.concat(
    [df_up, df_down],
    ignore_index=True
)
combined
```

#### merge the Ensembl biomart data

```python
df_ensembl = pd.read_csv("../data/biomart_goslim/biomart_rice_goslim_R56.tsv", sep="\t")
df_ensembl.rename(columns={'start1 (bp)': 'Gene start(bp)'}, inplace=True)
df_ensembl.rename(columns={'end1 (bp)': 'Gene end(bp)'}, inplace=True)
df_ensembl
```

```python
# 1. merge all data
merged_df = pd.merge(df_ensembl,combined,
                        left_on="Gene stable ID",
                        right_on="GENEID",
                     how="inner")

merged_df.to_csv("../data/rice_annotation/rice_annotation.tsv", sep="\t")
merged_df
```

```python
# 2. merge UP data
merged_df_up = pd.merge(df_ensembl,df_up,
                        left_on="Gene stable ID",
                        right_on="GENEID",
                     how="inner")

merged_df_up.to_csv("../data/rice_annotation/rice_annotation_up.tsv", sep="\t")
merged_df_up
```

```python
# 3. merge DOWN data
merged_df_down = pd.merge(df_ensembl,df_down,
                        left_on="Gene stable ID",
                        right_on="GENEID",
                     how="inner")

merged_df_down.to_csv("../data/rice_annotation/rice_annotation_down.tsv", sep="\t")
merged_df_down
```
